```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l2
from keras.regularizers import l1
from keras.callbacks import EarlyStopping
from keras.layers import Dropout
from sklearn.metrics import recall_score,f1_score,precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# load the dataset
dataset = pd.read_csv('diabetes.csv', delimiter=',')
```

```python
# split into input (X) and output (y) variables
X = dataset.drop("Outcome",axis = 1)
y = dataset["Outcome"]
```

```python
# test train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#convert target value
y_train_new = []
for i in y_train:
  if i == 1:
    y_train_new.append([1,0])
  else:
    y_train_new.append([0,1])
y_train_new = np.array(y_train_new) 
```

```python
y_test_new = []
for i in y_test:
  if i == 1:
    y_test_new.append([1,0])
  else:
    y_test_new.append([0,1])
y_test_new = np.array(y_test_new) 
```

## L2 REGULARIZATION

```python
# define the keras model with l2 regularization (alpha = 0.0001)
model_1 = Sequential()
model_1.add(Dense(500, input_dim=8, activation='sigmoid',kernel_regularizer=l2(0.0001),bias_regularizer=l2(0.0001),activity_regularizer=l2(0.0001)))
model_1.add(Dense(100, activation='sigmoid',kernel_regularizer=l2(0.0001),bias_regularizer=l2(0.0001),activity_regularizer=l2(0.0001)))
model_1.add(Dense(2, activation='softmax'))
```

```python
# compile the model
model_1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the model on the dataset
model_1.fit(X_train, y_train_new, epochs=1000, batch_size=70, verbose= 1) 
```

```python
# predict result
result = model_1.predict_classes(X_test)
y_test_conv=np.argmax(y_test_new, axis=1)
```

```python
# create confusion matrix
confusion_matrix(y_test_conv, result)
```

```python
# calculate accuracy, recall, precision, f score
accuracy = accuracy_score(y_test_conv, result)
print('Accuracy: %.3f' % accuracy)
recall = recall_score(y_test_conv, result)
print('Recall: %.3f' % recall)
precision = precision_score(y_test_conv, result)
print('Precision: %.3f' % precision)
score = f1_score(y_test_conv, result)
print('F-Measure: %.3f' % score)
```

## EARLY STOPPING REGULARIZATION

```python
#setting up early stopping function
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=50)
```

```python
#create model
model_3 = Sequential()
model_3.add(Dense(500, input_dim=8, activation='sigmoid'))
model_3.add(Dense(100, activation='sigmoid'))
model_3.add(Dense(2, activation='softmax'))
```

```python
# compile the keras model
model_3.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model_3.fit(X_train, y_train_new,validation_split=0.3, epochs=1000, batch_size=70, verbose= 1,callbacks=[es])
```

```python
# predicting result
result = model_3.predict_classes(X_test)
y_test_conv=np.argmax(y_test_new, axis=1)
```
